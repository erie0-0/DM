```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris,load_wine
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
iris=load_iris()
wine=load_wine()
```

```python
iris
```

```python
df= pd.DataFrame(data=iris.data ,columns=iris.feature_names)
df
```

```python
df.mean()
```

```python
df.std()
```

```python
#normalization
min_max_scaler= MinMaxScaler()
normalized= min_max_scaler.fit_transform(df)
print(normalized)
```

```python
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)
print(scaled_data)
```

```python
print(scaled_data.mean())
print(scaled_data.std())
```

```python
w_df= pd.DataFrame(data=wine.data ,columns=wine.feature_names)
w_df
```

```python
w_df.isna().sum()
```

```python
w_df.mean()
```

```python
w_df.std().all
```

```python
#normalization
min_max_scaler= MinMaxScaler()
normalized_data= min_max_scaler.fit_transform(w_df)
print(normalized_data)
```

```python
scaler=StandardScaler()
scaled_data=scaler.fit_transform(w_df)
print(scaled_data)
```

```python
print(scaled_data.mean())
print(scaled_data.std())
```
