```python
import pandas as pd
import numpy as np

data= pd.read_csv("groceriesDataset.csv")
data
```

```python
df = pd.DataFrame(data)
df
```

```python
dff = df.drop("Items", axis = 1)
dff
```

```python
li= []
for index, row in dff.iterrows() :
    li.append([item for item in row if pd.notna(item)])
print(li)
```

```python
data.describe
```

```python
from efficient_apriori import apriori
itemsets, rules = apriori(li, min_support = 0.005, min_confidence = 0.075)
itemsets
```

```python
rules
```

```python
frequent_itemsets = []
for size, itemset in itemsets.items():
    for items, support in itemset.items():
        frequent_itemsets.append((items,support))
frequent_itemsets
```

```python
sorted_i = sorted(frequent_itemsets, key = lambda x: x[1], reverse = True)
sorted_i
```

```python
for items, support in sorted_i:
    print(f"ITEMS: {','.join(items)} - Support :{support}")
```

```python
top_5 = sorted_i[:5]
top_5
```

```python
import matplotlib.pyplot as plt
item_names = [",". join(items)for items, _ in top_5]
support_val= [support for _, support in top_5]
plt.figure(figsize=)
plt.bar(item_names, support_val, color ="red")
plt.xlabel("Itemsets")
plt.ylabel("Support")
plt.show()
```
