```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris.target 
iris.data
```

```python
print("Feature names:", iris.feature_names)
print("Target names:", iris.target_names)
```

```python
data = pd.DataFrame(iris.data, columns = ['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)'])
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
data.drop_duplicates()
```

```python
plt.scatter(data.iloc[:,0], data.iloc[:,2])
```

```python
Kmeans = KMeans(n_clusters = 3, random_state = 0)
```

```python
Y = Kmeans.fit_predict(data[['sepal length (cm)','petal length (cm)']])
```

```python
Y
```

```python
Kmeans.labels_
```

```python
Kmeans.cluster_centers_
```

```python
fig, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 3], c=Y)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[3])
fig= ax.legend( scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes")
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
sc.fit_transform(data)
```

```python
from sklearn.cluster import  AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters = 3)
y_pred = agg.fit_predict(data)
y_pred
```

```python
fig, ax = plt.subplots()
scatter = ax.scatter(data.iloc[:, 0], data.iloc[:, 3], c= y_pred)
ax.set(xlabel= "Sepal Length", ylabel="Petal Width")
fig= ax.legend( scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes")
```

```python

from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
plt.figure(figsize=(18,10))
Z = linkage(data[['sepal length (cm)','petal length (cm)']], method='ward')
dendrogram(Z)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import pairwise_distances
```

```python
distance = pairwise_distances(data)
sorted_dist = np.sort(distance.ravel())
plt.plot(sorted_dist)
plt.xlabel("Data Points")
plt.ylabel("Distance")
```

```python
eps = 0.5
min_point = 5
db = DBSCAN(eps = eps, min_samples = min_point, metric = 'euclidean' )
```

```python
y_p = db.fit_predict(data)
```

```python
y_p
```

```python
fig, ax = plt.subplots()
scatter = ax.scatter(data.iloc[:, 0], data.iloc[:, 3], c= y_p)
ax.set(xlabel= "Sepal Length", ylabel="Petal Width")
fig= ax.legend( scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes")
```
